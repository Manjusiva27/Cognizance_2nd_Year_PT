```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame from the dataset
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target variable (species) to the DataFrame
df['species'] = iris.target

# Display the first 5 rows
df.head()
# Summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

# Plot histograms for each feature
df.hist(bins=20, figsize=(10, 8))
plt.show()
```

```python
import seaborn as sns

# Visualize pairwise relationships
sns.pairplot(df, hue='species', diag_kind='hist')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the features
scaled_data = scaler.fit_transform(df.iloc[:, :-1])

# Convert to a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=iris.feature_names)
scaled_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
minmax_scaler = MinMaxScaler()

# Fit and transform the features
minmax_scaled_data = minmax_scaler.fit_transform(df.iloc[:, :-1])

# Convert to a DataFrame
minmax_scaled_df = pd.DataFrame(minmax_scaled_data, columns=iris.feature_names)
minmax_scaled_df.head()
```
